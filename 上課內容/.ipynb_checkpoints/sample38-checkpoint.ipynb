```python
import pandas as pd
xlsx = pd.ExcelFile('總統.xls')
all=pd.read_excel(xlsx,'中　央')
all
```

```python
df=all.iloc[5:,:4]
df
```

```python
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']
```

```python
df
```

```python
df=df.set_index('縣市')
```

```python
df.T
```

```python
df
```

```python
#總得標
se1=df.sum(axis=0)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
figure, ax1 = plt.subplots()
se1.plot(ax=ax1,kind="bar",color=['b', 'g', 'orange'])
```

```python
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

ax1.set_xticklabels(['朱立倫','蔡英文','宋楚瑜'],fontproperties=font, rotation=360)
ax1.set_xlabel('選舉候選人',fontproperties=font)
ax1.set_yticklabels(['0','1佰','2佰','3佰','4佰','5佰','6佰','7佰'],fontproperties=font)
for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    print(x, y, width, height)
    ax1.annotate('{:.0f}票'.format(height), (x+width/4,height+100000), fontproperties=font)
```

```python
se2=se1/se1.sum() * 100
se2
```

```python
figure2, ax2 = plt.subplots()
se2.plot(ax=ax2,kind="bar",color=['b', 'g', 'orange'])

ax2.set_xticklabels(['朱立倫','蔡英文','宋楚瑜'],fontproperties=font, rotation=360)
ax2.set_xlabel('選舉候選人',fontproperties=font)
#ax2.set_yticklabels(['0','1佰','2佰','3佰','4佰','5佰','6佰','7佰'],fontproperties=font)
for p in ax2.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    print(x, y, width, height)
    ax2.annotate('{:.2f}%'.format(height), (x+width/4,height), fontproperties=font)
```

```python
def do1(x):
    total = x['朱立倫'] + x['蔡英文'] + x['宋楚瑜']
    x['朱立倫'] = x['朱立倫'] / total
    x['蔡英文'] = x['蔡英文'] / total
    x['宋楚瑜'] = x['宋楚瑜'] / total
    return x
    

df1=df.apply(do1,axis=1);
df1.sort_values('蔡英文',ascending=False)
```
